```python
from PIL import Image
import os
import cv2

def convert(dir,width,height):
    file_list = os.listdir(dir)
    # print(file_list)
    for filename in file_list:
        # try:
        path = dir+ "/" +filename
        im = Image.open(path)
        out = im.resize((width,height),Image.ANTIALIAS)
        # except:
        #     os.remove(path)
    # print("%s has been resized!"%file_list)

def img_check(dir):
    file_list = os.listdir(dir)
    # print(file_list)
    for filename in file_list:
        # try:
        path = dir+ "/" +filename
        # print(path)
        img = cv2.imread(path)
        try:
            if img.shape[2] != 3:
                pass
        except:
            print(filename, "False!!!!")
            os.remove(path)

if __name__ == '__main__':
   
    dataset_dir = "dataset/images"
    all_filedirs = os.listdir(dataset_dir)
    for class_folder in all_filedirs:
        img_dir = dataset_dir + "/" + class_folder
        convert(img_dir,256,256)
        img_check(img_dir)
    print("process done!")
```
